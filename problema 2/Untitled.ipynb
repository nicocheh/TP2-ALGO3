```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math

df2 = pd.read_csv('tiemposMFijos.csv', ' ')
for i in range (1,100,1):
    df2[df2['cantServidores']==i].plot.scatter(x='cantEnlaces', y='tiempoConsultora2', color='Blue', label='Mfijo, consultora 2')
    df2[df2['cantServidores']==i].plot.scatter(x='cantEnlaces', y='tiempoConsultora1', color='Blue', label='Mfijo, consultora 1')

    plt.ylabel('Segundos')
    plt.xlabel('Enlaces')
    plt.show()
```
