```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math

df2 = pd.read_csv('instanciasRandom2.csv', ' ')

df2.plot.scatter(x='clientes', y='tiempo', color='Blue', label='Random')

plt.ylabel('Ticks')
plt.xlabel('Clientes')
plt.show()

df2['tiempo']=(df2['tiempo']/df2['clientes'])/df2['clientes']

df2.plot.scatter(x='clientes', y='tiempo', color='Blue', label='Random')

plt.ylabel('Ticks')
plt.xlabel('Clientes')
plt.show()

df2.plot.scatter(x='clientes', y='tiempo', color='Blue', label='Random')

plt.ylabel('Ticks')
plt.xlim(20,60)
plt.xlabel('Clientes')
plt.show()
```
