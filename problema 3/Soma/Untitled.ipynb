```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
import math




```

```python
df = pd.read_csv('instanciasRandom2.csv', ' ')
df2 = pd.read_csv('mejorcaso.csv', ' ')
df3 = pd.read_csv('peorcaso.csv', ' ')
```

```python

df.plot.scatter(x='clientes', y='tiempo', color='Blue', label='Random')

plt.ylabel('Ticks')
plt.xlabel('Clientes')
plt.show()



df2.plot.scatter(x='clientes', y='tiempo', color='Green', label='Mejor')

plt.ylabel('Ticks')
plt.xlabel('Clientes')
plt.show()




df3.plot.scatter(x='clientes', y='tiempo', color='Red', label='Peor')

plt.ylabel('Ticks')

plt.xlabel('Clientes')
plt.show()

ax= df.plot.scatter(x='clientes', y='tiempo', color='Blue', label='Random')
df2.plot.scatter(x='clientes', y='tiempo', color='Green', ax=ax, label='Mejor')

df3.plot.scatter(x='clientes', y='tiempo', color='Red', ax=ax, label='Peor')
plt.xlim(0,60)
plt.ylabel('Ticks')
plt.xlabel('Clientes')

plt.show()
```
